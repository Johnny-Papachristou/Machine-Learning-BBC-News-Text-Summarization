```python
df = pd.read_csv('bbc_text_cls.csv')
```

```python
df.head()
```

```python
doc = df[df.labels == 'business']['text'].sample(random_state=42)
```

```python
def wrap(x):
  return textwrap.fill(x, replace_whitespace=False, fix_sentence_endings=True)
```

```python
# Split to sentences
sentences = nltk.sent_tokenize(doc.iloc[0].split("\n", 1)[1])
print(len(sentences))
```

```python
# Compute TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(sentences)
print(X.shape)
```

```python
# Compute scores
non_zero_averages = []

# Iterate through each row in the TF-IDF matrix
for i in range(X.shape[0]):
    row = X[i]
    # Get the non-zero TF-IDF scores
    non_zero_scores = row.data
    # Calculate the average of the non-zero scores
    if non_zero_scores.size > 0:
        avg_score = np.mean(non_zero_scores)
    else:
        avg_score = 0  # Handle case with no non-zero scores
    non_zero_averages.append(avg_score)

# Store the averages in a new variable
average_tfidf_scores = np.array(non_zero_averages)

# Print the results
print("Average TF-IDF scores for each sentence:")
print(average_tfidf_scores)
```

```python
# Sort by highest score
sorted_indices = np.argsort(average_tfidf_scores)[::-1]

# Print the sorted sentences
for i in sorted_indices:
    print(sentences[i])
```
